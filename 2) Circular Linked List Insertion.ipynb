```python
# Circular Linked List Insertion if Link is empty :-

class Node:
    def __init__(self, data):
        self.data = data
        self.next = None
        
class CircularLinkedList:
    def __init__(self):
        self.head = None
        self.tail = None
        
    def display(self):
        if self.head is None:
            print("The CLL is empty")
        else:
            
            temp = self.head
            print(temp.data, "-->", end=" ")
            
            while temp.next!=self.head:
                temp = temp.next
                print(temp.data, "-->", end=" ")
            print(temp.next.data)    
            
    def add_beginning(self,data):
        new = Node(data)
        
        if self.head is None:
            self.head = new
            self.tail = new
            self.tail.next = self.head
            
        else:
            new.next = self.head
            self.tail.next = new
            self.head = new
            
    def add_ending(self, data):
        new = Node(data)
        
        if self.head is None:
            self.head = new
            self.tail = new
            self.tail.next = self.head 
        else:
            self.tail.next = new
            self.tail = new
            self.tail.next = self.head
            
    def add_element_specified_position(self,pos,data):
        new = Node(data)
        if self.head is None:
            self.head = new
            self.tail = new
            self.tail.next = self.head
            
        else:
            temp = self.head 
            for i in range(1,pos-1):
                temp = temp.next
            new.next = temp.next
            temp.next = new
            
        
l = CircularLinkedList()

l.add_beginning(10)
print("Insertion at beginning")
l.display()

l.add_beginning(5) # You can keep adding as many you want
print("Insertion at beginning")
l.display()

l.add_beginning(4)
print("Insertion at beginning")
l.display()

l.add_ending(15)
print("\nInsertion at ending")
l.display()

l.add_ending(20)
print("Insertion at ending")
l.display()

l.add_element_specified_position(4,14)
print("\nInsertion at specified position")
l.display()

l.add_element_specified_position(6,16)
print("\nInsertion at specified position")
l.display()
```
